```python
from os import getcwd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from fish_simulator.simulator import (
    run,
    plot_skeletal_postures_with_trace,
    plot_tail_image_with_trace,
    plot_skeletal_postures
)

PLOT_URL = ("https://gist.githubusercontent.com/"
       "ThomasMullen/e4a6a0abd54ba430adc4ffb8b8675520/"
       "raw/1189fbee1d3335284ec5cd7b5d071c3da49ad0f4/"
       "figure_style.mplstyle")
plt.style.use(PLOT_URL)
```

```python
# load tail angle data
data_name = "swim01"
data_arr = np.load(Path(Path(getcwd()).parent, "test", "fixtures", f"{data_name}.npy"))

# plot tail angle data
fig, ax = plt.subplots(figsize=(3.5,3.5), dpi=300)
ax.plot(data_arr)
ax.set(xlabel="Time (frames)", 
       ylabel="Tail angle (Radians)",
       title="Tail trace")
fig.tight_layout()
```

```python
# define a directory to save the plots and video
fig_dir = Path(Path(getcwd()).parent, "dump", "plts")


# Plot the posture on an animated fish structure
run(
    data_arr,
    plot_func=plot_skeletal_postures,
    png_dir=f"{fig_dir}",
    vid_fp=f"{fig_dir.parent}/run_ani_{data_name}_type2.mp4",
    n_intp_segs=30,
    line_wid=1,
)
```
